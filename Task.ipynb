```python
import pandas as pd
import numpy as np
```

```python
#Create a 0 dimension, 1 dimension and 2 dimension numpy array
# 0 dimension array
arr0=123
print(arr0)
# 1 dimension array
arr1=[1,2,3]
print(arr1)
# 2 dimension array
arr2=[[1,2,3],[4,5,6]]
print(arr2)
```

```python
#- Create a 2 dimension array and

arr=[[1,2,3],[4,5,6]]
print(arr1)

print("Sum of array elements:", np.sum(arr))
print("Minimum value in array:", np.min(arr))
print("Maximum value in array:", np.max(arr))
print("Mean of array elements:", np.mean(arr))
print("Median of array elements:", np.median(arr))
print("Standard deviation of array elements:", np.std(arr))
```

```python
#Create a 1 dimension array using np.arange()
arr3d = np.arange(1,13)
print("1D array:\n",arr3d)

#Then restructure that same array using np.reshape() to 2 dimension array.
arr4d = arr3d.reshape(6, 2)
print("reshape:",arr4d)

# Transpose that array to opposite of the same reshaped array.
arr5d = arr4d.transpose()
print("transpose:",arr5d)

#Flatten the same array to normal array that we created first.
arr6d=arr5d.flatten()
print("flatten:",arr5d)
```

```python
#Create a 2 dimension array using nested list
arr7d=np.array([[10,11,12],[4,5,6],[7,8,9]])
print(arr7d)

# use np.sort() to sort the array
arr10=np.sort(arr7d)
print("sorted array\n:",arr10)

#use np.sort()[::-1] to reverse the sorted array.
sorting=np.sort(arr7d)[::-1]
print("reverse\n :",sorting)

#use np.size() to check the size and np.shape() to check the shape of the array
sized=np.size(arr7d)
print("size:\n",sized)
```

```python
arr2d = np.array([[1,2,3,4],
                  [4,5,6,7],
                  [7,8,9,10]])
# Use this array and access 1st last two elements from the array (9,10)
print(arr2d[2, 2:3])
# Use this array and access only last column elements (4,7,10)
print(arr2d[-1, -2:])
```
